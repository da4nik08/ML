```python
def EqualDiscret(n_part, name_atr, data):
    
    def transf(x):
        min1 = data[name_atr].min()
        max1 = data[name_atr].max()
        return (x - min1) // ((max1 - min1) / n_part) if x < max1 else (n_part - 1)
    
    data[name_atr] = data[name_atr].transform(transf)
```

```python
def EuclidianDist(dataset, inst, target):
    distances = []
    
    for i, row in dataset.iterrows():
        temp = 0
        
        for j in dataset.columns.values:
            if j != target:
                temp += (row[j] - inst[j]) ** 2
        
        distances.append((math.sqrt(temp), str(row[target]))) 
        
    return distances
```

```python
def KnnOne(dataset, target, instance, k, discrete=True):
    dist = EuclidianDist(dataset, instance, target)
    dist.sort()
    neigh = dist[0:k]
    targets = [target for (dist, target) in neigh]
    amount = collections.Counter(targets)
    return max(amount, key=amount.get)
```

```python
def knn_simple_vote(observation, data_segmentation, k, target):
    predictions = [KnnOne(data, target, observation, k) for data in data_segmentation]
    frequency = collections.Counter(predictions)
    return max(frequency, key=frequency.get)
```

```python
def knn_weighted_vote(weights, observation, data_segmentation, k, target):
    predictions = [KnnOne(data, target, observation, k) for data in data_segmentation]
    probabilities = {}
    for i in range(len(predictions)):
        probabilities[predictions[i]] = probabilities.get(predictions[i], 0.) + weights[i]
    return max(probabilities, key=probabilities.get)
```

```python
def knn_dynamic_vote(data_train: pd.DataFrame, data_test: pd.DataFrame, k, target):
    weights = [1 * i for i in range(len(data_train))] 
    results = []
    for index, observation in data_test.iterrows():
        predictions = [KnnOne(data, target, observation, k) for data in data_train]
        probabilities = {}
        for i in range(len(predictions)):
            probabilities[predictions[i]] = probabilities.get(predictions[i], 0.) + weights[i]
        result = max(probabilities, key=probabilities.get)
        for i in range(len(predictions)):
            if predictions[i] != observation[target]:
                weights[i] /= 2
        results.append(result)
    return results
```

```python
def get_cross_validation_train(data: pd.DataFrame):
    kf = KFold(n_splits=10, random_state=52, shuffle=True)
    data_train, data_test = [], []
    for train_index, test_index in kf.split(data):
        data_train.append(data.iloc[train_index])
        data_test.append(data.iloc[test_index])
    return data_train, data_test
```

```python
def get_weights(data_train: pd.DataFrame, data_test: pd.DataFrame, target): 
    return [accuracy_score(data_test[target].to_list(), [KnnOne(data, target, observation, 3) 
            for index, observation in data_test.iterrows()]) for data in data_train]
```

```python
def get_cross_validation_weights(data_train: pd.DataFrame, data_test: pd.DataFrame, target):
    return [accuracy_score(data_test[i][target].to_list(), [KnnOne(data_train[i], target, observation, 3) 
        for index, observation in data_test[i].iterrows()]) for i in range(len(data_train))]
```
