```python
def EqualDiscret(n_part, name_atr, data):
    
    def transf(x):
        min1 = data[name_atr].min()
        max1 = data[name_atr].max()
        return (x - min1) // ((max1 - min1) / n_part) if x < max1 else (n_part - 1)
    
    data[name_atr] = data[name_atr].transform(transf)
    
```

```python
def CountClass(dataset, name_col):
    output = dict()
    for i in dataset[name_col]:
        if i in output:
            output[i] += 1
        else:
            output[i] = 1
    return output        
```

```python
def InformGain(arr, name, s_data, class_name):
    
    size = s_data.shape[0]
    
    class_data = CountClass(s_data, class_name)
    
    output = list()
    
    data = list()
    
    d = sum(list(class_data.values()))
    
    for value in class_data.values():
        data.append(value/d)
    
    entD = entropy(data, base=len(data))
    
    for i in arr:
        
        less = s_data[:int(i)]
        
        more = s_data[int(i):]
        
        lower = CountClass(less, class_name)
        
        higher = CountClass(more, class_name)
        
        d_lower = sum(list(lower.values()))
        
        d_higher = sum(list(higher.values()))
        
        data_lower = list()
        
        for value in lower.values():
            data_lower.append(value/d_lower)
        
        data_higher = list()
        
        for value in higher.values():
            data_higher.append(value/d_higher)
        
        if(len(data_lower) > 1):
            entL = entropy(data_lower, base=len(data_lower))
        else:
            entL = 0
            
        if(len(data_higher) > 1):
            entH = entropy(data_higher, base=len(data_higher))
        else:
            entH = 0
        
        output.append(entD - ((d_lower / d) * entL + (d_higher / d) * entH))
        
    return output
```

```python
def InformGainDiscret(name_atr, class_name,  data_iGc, n_part=3):
    s_data = data_iGc.sort_values(by=[name_atr])
    arr = GetÐ¡hange(s_data, name_atr)
    gain = InformGain(arr, name_atr, s_data, class_name)
    gain, arr = zip(*sorted(zip(gain, arr), reverse=True))
    
    l = list(arr[:n_part])
    l.sort()
    l.insert(0, 0)
    idx = list()
    
    for i in l:
        idx.append(list(s_data.index)[int(i)])
    for i in range(1, len(l)):
        s_data.loc[int(idx[i-1]): int(idx[i]), name_atr] = i - 1
    s_data.loc[int(idx[-1]):, name_atr] = n_part
    return s_data
    
```

```python
class Node:
    next_n = dict()
    data = None
    name_target = None
    current_name = None
    def __init__(self, data, name_target):
        self.data = data
        self.name_target = name_target
        
    def CountClass(self, name_col):
        output = dict()
        for i in self.data[name_col]:
            if i in output:
                output[i] += 1
            else:
                output[i] = 1
        return output
    
    def CountClass2(self, data, name_col):
        output = dict()
        for i in data[name_col]:
            if i in output:
                output[i] += 1
            else:
                output[i] = 1
        return output
    
    def InformGain(self):
    
        #size = self.data.shape[0]
    
        class_data = self.CountClass(self.name_target)
    
        output = list()
    
        data = list()
    
        d = sum(list(class_data.values()))
    
        for value in class_data.values():
            if int(value) != 0:
                data.append(value/d)
       
        if len(data) > 1:
            entD = entropy(data, base=len(data))
        else:
            entD = 0
    
        columns = list()
        
        attr = dict()
    
        for i in self.data:
            
            if i != self.name_target:
                columns.append(i)
            
        for i in columns:
            
            attr[i] = self.CountClass(i)
            
        all_gain = dict()
        
        for at in attr:
            
            ent = entD
            for cls in attr[at]:
                
                filter = self.data[at] == cls
                temp = self.data.where(filter)
                temp = temp.dropna()
                
                class_data = self.CountClass2(temp, self.name_target)
    
                output = list()
    
                data = list()
    
                d_temp = sum(list(class_data.values()))
    
                for value in class_data.values():
                    if int(value) != 0:
                        data.append(value/d_temp)
                if len(data) > 1:
                    ent = ent - ((d_temp/d) * entropy(data, base=len(data)))
                
                    
            all_gain[at] = ent
            
        return all_gain
    
    def update(self):
        
        classes = self.CountClass(self.name_target)
        if len(classes) <= 1:
            self.next_n = str(list(classes.keys())[0])
            return
        
        if self.data.shape[1] <= 1:
            self.next_n = str(max(classes, key=classes.get))
            return 
        
        inf_gain = self.InformGain()
        self.current_name = max(inf_gain, key=inf_gain.get)
        
        cur_classes = self.CountClass(self.current_name)
        
        for key in cur_classes:
            filter = self.data[self.current_name] == key
            
            temp_d = self.data.where(filter)
            temp_d = temp_d.dropna()
            
            self.next_n[key] = Node(temp_d.drop(self.current_name, axis='columns'), self.name_target)
            self.next_n[key].update()
    
    def step(self, data):
        if isinstance(self.next_n, str):
            return self.next_n
        attr = data[self.current_name]
        return self.next_n[attr].step(data)
```

```python
class ID3Tree:
    root = None
    def __init__(self, data_set, target):
        self.root = Node(data_set, target)
    
    def Build(self):
        self.root.update()
        
    def Predict(self, data):
        output = list()
        
        for i in range(data.shape[0]):
            output.append(self.root.step(data.iloc[i]))
            
        return output
```
